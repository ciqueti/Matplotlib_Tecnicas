```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots(figsize=(15,10))
ax.plot([2,6,12,20], marker='s')
ax.grid(True)
ax.set_xlabel('Eixo X', loc='left')
ax.set_ylabel('Eixo Y', loc='top')
ax.set_title('Testando Gráficos',fontname='serif', loc='left', color='red' , fontsize=15)
ax.legend(['Linha'])
```
